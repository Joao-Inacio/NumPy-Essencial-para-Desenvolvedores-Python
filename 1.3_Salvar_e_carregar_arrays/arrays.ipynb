# **Salvando e Carregando arrays**

```python
import numpy as np
```

```python
array_1 = np.array([0, 1, 2, 3, 4])
array_2 = np.array([5, 6, 7, 8, 9])
```

## np.save e np.load

```python
np.save(file='array_1_file', arr=array_1)
```

```python
array_1_loaded = np.load(file='./array_1_file.npy')
array_1_loaded
```

## np.savez

```python
np.savez('array_savez', array_1, array_2)
```

```python
array_loaded = np.load('./array_savez.npz')
array_loaded
```

```python
array_loaded['arr_0']
```

```python
array_loaded['arr_1']
```

## np.savez_compressed

```python
np.savez_compressed('arrays_compacted', array_1, array_2)
```

```python
array_compacted_loaded = np.load('./arrays_compacted.npz')
array_compacted_loaded
```

```python
array_compacted_loaded["arr_0"], array_compacted_loaded["arr_1"]
```

## np.savetxt e np.loadtxt

```python
array_2d = np.vstack([array_1, array_2])
array_2d
```

```python
np.savetxt(fname='array_csv.csv', X=array_2d)
```

```python
np.savetxt(fname='array_csv.csv', X=array_2d, fmt='%s')
```

```python
np.savetxt(fname='array_csv.csv', X=array_2d, fmt='%s', delimiter=',')
```

```python
array_csv_loaded = np.loadtxt(fname='./array_csv.csv', dtype=int, delimiter=',')
array_csv_loaded
```
